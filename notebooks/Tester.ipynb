```python
big_data_names = pd.concat([big_data['R_fighter'], big_data['B_fighter']]).unique()
small_data_names = small_data['fighter_name'].unique()

# Step 2: Create a mapping of names to their order in small_data
name_order = {name: index for index, name in enumerate(small_data_names)}

# Step 3: Add an order column to big_data based on the small_data order
big_data['order'] = big_data['R_fighter'].map(name_order).fillna(-1)  # Use -1 for names not found
big_data['order'] = big_data['order'].where(big_data['order'] != -1, big_data['B_fighter'].map(name_order).fillna(-1))

# Step 4: Sort big_data based on the order column
big_data_sorted = big_data[big_data['order'] != -1].sort_values('order').drop('order', axis=1)

# Display the sorted big_data
print(big_data_sorted)
```

```python
columns_to_drop = big_data_sorted.filter(like='B').columns

# Step 2: Drop those columns
big_data_sorted= big_data_sorted.drop(columns=columns_to_drop)
```

```python
big_data_sorted.drop(columns=['Referee'], inplace=True)
```

```python
big_data_sorted = big_data_sorted.loc[:, ~big_data_sorted.columns.str.contains('opp')]
big_data_sorted.info()
```

```python
big_data_sorted.isna().sum()
```

```python
# Select numerical columns from the DataFrame
numerical_columns = big_data_sorted.select_dtypes(include=['float64', 'int64']).columns

# Step 1: Group by 'R_fighter' and calculate the mean for all numerical columns
fighter_mean_values = big_data_sorted.groupby('R_fighter')[numerical_columns].transform('mean')

# Step 2: Fill missing values in all numerical columns with the corresponding fighter's mean value
big_data_sorted[numerical_columns] = big_data_sorted[numerical_columns].fillna(fighter_mean_values)

# Check the result
big_data_sorted.isna().sum()
```

```python
# Drop duplicate rows based on the 'R_fighter' column, keeping only the first occurrence
big_data_filtered = big_data_sorted.drop_duplicates(subset='R_fighter', keep='first')
big_data_filtered.info()
```

```python
big_data_filtered.head()
```

```python
# Reset the index and drop the old index
big_data_filtered = big_data_filtered.reset_index(drop=True)
```

```python
# Merge on 'fighter_name' from small_data and 'R_fighter' from big_data_filtered
merged_data = pd.merge(small_data, big_data_filtered, left_on='fighter_name', right_on='R_fighter', how='inner')

# Optionally, drop the 'R_fighter' column if it's redundant after merge
merged_data = merged_data.drop(columns=['R_fighter'])

# Check the result
merged_data.head()
```
